## Hijack: the game

**Game description**

With some plot twist (I hope) that will keep you on the edge of your seat ( I reeeeally hope so), the user is going to go through the hijacking of flight KA29 and try to end it as safely as possible for all on board. However, as the user moves from stage, the decisions become more complicated. Are your negotiation skills and luck on your side?

```python
#imports
import time

#lists
yes_lst = ["yes", "sure", "of course", "okay", "why not", "ok", "k", "absolutely", "certainly", "definitely", "indeed", "affirmative", "agreed", "roger", "aye", "aye aye", "aye aye captain", "fine", "alright", "positive", "yep", "yeah", "yah", "yea", "sure thing"]
no_lst = ["no", "nope", "not at all", "negative", "never", "absolutely not", "refuse", "decline", "deny", "not really", "not sure", "unlikely", "doubtful", "impossible", "I disagree", "I object", "I dissent", "veto", "forbidden", "prohibited", "not permitted", "not allowed", "restricted"]
easter_egg = ["why", "why?", "Why", "Why?", "what for?", "what for", "What for","What for?"]
failure_messages = {"dubai": "\nFlight KA29, lost. The hijackers gain aircraft's control. You lose.",
                    "phone_access":"""
        You try to hide your phone from the hijackers, but they catch you.
        Instead of taking it away from you, the knock you out.
        
        The flight has been fully taken over and there is no one left to challenge the hijackers, you lost.
        """, "hijackers" :"\nFlight KA29, lost. The hijackers kept absolute control of the situation. You lose.", "gun": """The other hijackers attack you, you lose gun control and the hijackers are now more agressive.
they continue without no one to confront them anymore.""", "pilot_sam": """\n You have nor paper nor pen... your plan cannot be and the hijackers find you weird so they
lock you on the bathroom to avoid further distractions. you lose.""","air_control": "Flight KA29, lost. Iraq's government shot the plane due to national security."  }

win_messages = {"families_no": """Flight KA29, incommunicated and on air. The situation is uncertain but, 
                                          we are doing everything in our power to maintain the situation under control.
                                          \n        Hijack 2    Coming soon""",
                "families_yes": "\nThe hijacking ended but the hijackers and Sam Nelson's families were killed.\n        Hijack 2    Coming soon",
               "stop_plane": """\nSam talked to the copilot so she could explain in hungarian
                    the situation. While the exchange happened, Terry (bullet guy) told him the hijacker's master 
                    mind's reach and 'ethics'. He realized the magnitude of the situation. The are not just against the people 
                    on board.\n
                    The hijackers families and, Sam's family's life are at stake.""" }

def intro():
    response = input("""Sam Nelson arrives at Dubai International Airport after closing a tough negotiation. 
Now, he is flying back to London to reunite with his son and, hopefully, reconnect with his ex-wife. 
Right after his boarding pass check, a man barely reached the counter. 
The plane was closing so the crew were denying him entry. 
Do you help him? (Yes or No)\n""").lower()
    return validateInput(response)

def omniscientStart():
    print( """\nThe crew start his boarding pass check while Sam gets to the plane. 
    He then texted Marsha, his ex, 'I'm coming home'""")
    time.sleep(1.7)
    print( """\nNo answer and the plane takes off.""")
    time.sleep(1.5)
    print( """\nAs the flight progressed, a girl goes to the restroom and she noticed something; a bullet.
    Not knowing what to do, she rushed back to her seat and told her friends. The passanger behind, Terry, 
    noticed the girls and offered to take the bullet to the captain to asses the situation.
    They gave it to him and he walked towards the front of the plane. 
    But, he stopped at another passenger's seat, Stuart. They disscussed and decided to bring THE PLAN forward.
    Then, Terry went back to calm the girls who weren't satisfied with his explanation. So, they called a crew member.
    They soon realised he knew nothing about the rising issue. 
    Meanwhile, Terry knew that they had to act before the crew. And, in the blink of an eye...""")
    
    time.sleep(10)
    print( """     ...the plane was hijacked.""")
    
    dubai_result = dubaiAirControl()
    if dubai_result in yes_lst:
        hijackers()
    else:
        fail("dubai")

def samStart():
    print("""\nThe plane took off.\n \nAfter a while on air, you noticed a conversation between passangers that somehow, felt odd. 
Then you see washbags passing around passengers. But when requested, the crew told you they do not provide them or any similar items.
These raised some flags. You heard some commotion and then, someone took out a gun... """ )
    time.sleep(3)
    print( """      ...the plane was hijacked.\n""")
    time.sleep(1.2)
    
    phoneAccess()
    gunAction()
    pilotSam()

def hijackers():
    prompt = """\nThe hijackers knew about the pilot's affair with a crew member, Colette. 
They used her as a hostage and convinced the pilot to open the cockpit. Despite the fierce 
copilot's efforts to stop him, he managed to knock her out, open the cockpit, and told 
Dubai's air control the problem was resolved since they did not have specifics of it.
\nWhile all of this occurred at the front, the rest of the hijackers took everybody's electronics. 
Now, no one has communication with the outside.\n 
Sam Nelson was one of the few passengers who remained calm. And began to analyse the situation. 
	Talk to the hijackers?\n"""
    
    result = validateInput(input(prompt = prompt))
    
    if result in yes_lst:
        print("""\nAfter a failed attempt of getting a gun from the hijackers, Sam gained their trust 
by giving it back. He then convinced them to get the pilot out of the cockpit since the plane is on autopilot.
He is just as useful while air control contacts the plane. \n
They moved him to a business class seat, and the pilot is now reachable.\n
Both the pilot and Sam devised a plan to alert air control in the UK without raising the hijackers' suspicions;
adjust the flight plan by a few degrees.\n""")
        return
    fail("hijackers")

def iraq():
    time.sleep(2)
    prompt = """A few hours have passed and the passengers began to plot among each other to stop the hijacking. 
	Attack the hijackers?\n"""
    return validateInput(input(prompt= prompt))
    
    
def airControl():
    prompt = """\nThe passangers in Economy made another attempt against the hijackers. 
    Meanwhile, the plane is now crossing Iraq's air space. 
    \nThe aircraft need to answer Iraq's air control or else, they are going to shot down the plane since it will be
perceived as national security threat. 
With the revolt in the plane, the hijackers were dispersed and Stuart didn't allowed the pilot to answer the call.
	Talk to Iraq's air control?\n """
    
    return validateInput(input(prompt = prompt))
    
def stopPlane():
    prompt = """\nThe pilot, after several contact attempts, answered Iraq's call and successfully crossed  
their air space. They are now in Hungary, the copilot's birth country; a new oportunity to land the plane appeared.
                    \nAfter Stuart's brother condition deteriorates, Sam is trying to convince Stuart to stop just enough to get
his brother medical care.
	Stop the plane?\n"""
    
    
    result = validateInput(input(prompt = prompt))
    if result in yes_lst:
        families()
    else:
        win("stop_plane")

def families():
    prompt = """\nSam talked to the coplit so she could explain the situation in hungarian. 
While the exchange happened, Terry (bullet guy) told him the hijacker's master mind's reach and 'ethics'. He 
realized the magnitud of the situation. The are not just against the people on board.\n
The hijackers families and, Sam's family's life are on stake.
	Are you sure you want to stop the plane?\n"""
    result = validateInput(input(prompt = prompt))
    if result in yes_lst:
        win("families_yes")
        return
    win("families_no")
    

def phoneAccess():
    print("""The hijackers took a crew member as a hostage and convinced the pilot to open the cockpit. Despite the fierce copilot's efforts to 
stop him, he managed to knock her out, open the cockpit, and told Dubai's air control the problem was resolved 
since they did not have specifics of it. In the meantime...""")
    while True:
        try:
            phone = int(input(prompt= """ \nThe hijackers are asking everyone for their electronics, do you:
        1.- Hand over your phone
        2.- Hide your phone
        3.- Try to get a message out before they get to your place
        """))
            if phone == 1:
                print("\nYou immediately comply and hand over your phone to the hijackers.")
                return False
            elif phone == 3:
                print("""\nYou successfully sent a message to your ex wife about the current situation.
                When the hijackers ask for your phone, you just handed it.""")
                return True
            elif phone == 2:
                fail("phone_access")
            else:
                raise ValueError("\nThe option you selected does not match any of the options presented, please select a valid option")

        except ValueError as error:
            print(error)
            
            
def gunAction():
    print("""\nAs the situation intensified, passangers began to plot against the hijackers. 
You try to persude them but they did it anyways (people :/) and failed to take a gun from them.
Suddenly, it was on the floor. You grab it.\n""")
    while True:
        try:
            gun = int(input(prompt= """ 	What do you do with it?
1. Give it back 
2. Shoot a hijacker\n """))
            if gun == 1:
                print("""\nYou earned the hijackers 'trust'. Now that they listen to you; you talked them into 
believing that the pilot inside the cockpit is a problem since he is just worthy while talking to air control. 
So, they take him to a business class seat. And, he is now available.\n""" )
                return False
            elif gun == 2:
                fail("gun")
                break
            else:
                raise ValueError("\nThe option you selected does not match any of the options presented, please select a valid option")

        except ValueError as error:
            print(error) 
            
            
def pilotSam():
    while True:
        try:
            options = [
                "1. With the force.",
                "2. By seat's messaging",
                "3. Paper tossing"
            ]

            for index, option in enumerate(options, start=1):
                print(f"{option}")

            pilot_sam = int(input("\nHow do you approach the pilot?\n"))

            if pilot_sam == 1:
                print("\nAlthough the force is a great effort and power, it's not enough for this situation. You joined the dark side and evidently lose the game.")
                return False
            elif pilot_sam == 2:
                print("\nUsing a game between passengers, you and the pilot elaborate a plan to send a discreet distress signal. You only need the opportunity.")
                return True
            elif pilot_sam == 3:
                fail("pilot_sam")  # Assuming fail is a defined function
                return False
            else:
                raise ValueError("\nThe option you selected does not match any of the presented options. Please select a valid option.")

        except ValueError as error:
            print(error)  

def validateInput(response, easterEggEnabled = False):
    while response not in yes_lst and response not in no_lst:
        if easterEggEnabled and response in easter_egg:
            return response
        print("\nInvalid input, please try again")
        response = input('Yes or No? ').lower()
    return response

def dubaiAirControl():
    while True:
        prompt = """\nThe crew managed to inform the cockpit (where the pilots are) of the situation. 
        Do they inform aircontrol about the issue?\n"""
        result = validateInput(input(prompt = prompt), True)

        if result in easter_egg:
            print("""        After 9/11, changes were made to the security of cockpits in an effort to make hijackings 
            more difficult. According to the US Federal Aviation Administration, doors should typically be tough enough 
            to withstand a grenade blast. They are usually left locked throughout the flight""")
            replay = input(prompt="   Would you like to return to Dubai control? Press enter to continue ").lower()
        else:
            break
        
    return result
    
def fail(section):
    print("\n", failure_messages[section], "\n")
    raise SystemExit()
    
def win(section):
    print("\n", win_messages[section], "\n")
    raise SystemExit()
    
# Case control for input name
try:
    help_response = intro()
    if help_response in yes_lst:
        omniscientStart()
    elif help_response in no_lst:
        samStart()
    else:
        print("What are you like this?")

    iraq_response = iraq()
        if iraq_response in yes_lst:
            airControl()

        elif iraq_response in no_lst:
            print("The spirit of the passangers worsen by the minute. However, they afraid of the hijackers so everyone complied. ")
        else:
            fail("air_control")
    
    air_control_response = airControl()
        if air_control_response in yes_lst:
            stopPlane()
        else:
            fail("air_control")
    
    
except SystemExit:
    print("Game Over.")
```
